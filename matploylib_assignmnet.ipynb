# 1. Setup and Styles

```python
import matplotlib.pyplot as plt
import numpy as np
print(plt.style.available)
```

```python
plt.style.use('seaborn-v0_8-whitegrid') #setting default style as seaborn
```

# 2. Basic Line Plot

```python
x = np.linspace(-2,4) #set range
y = (x**2 -2*x +1)
plt.plot(x, y)
plt.show()
```

#3. Multiple Lines

```python
x = np.linspace(-2,4)
f = (x**3)
g = (np.e**x)
h = (np.log(x))
plt.plot(x,f , label='x**3')
plt.plot(x,g, label='e**x')
plt.plot(x,h, label = 'log(x)')
plt.legend()
plt.show()
```

# 4. Styling Lines

```python
x = np.linspace(-2,4)
f = (x**2)
g = (x**3)
h = (x**4)
i = (x**5)
plt.plot(x,f, color ='blue') # plotting and setting color by name
plt.plot(x,g, color = 'g') # plotting and setting color by short color code
plt.plot(x,h, color = '#9A318A') # plotting and setting color by hex code
plt.plot(x,i, color = '0.5') # plotting and setting number using greyscale
plt.show()
```

# 5. Axes Limits and Scaling

```python
x = np.linspace (-10,10)
y = (1/((x**2) + 1))
plt.plot(x,y)
plt.xlim(-10,10)
plt.ylim(-1,1)
plt.show()
```

```python
x = (1/((x**2) + 1))
y = np.linspace(-10,10)
plt.plot(x,y)
plt.xlim(-10,10)
plt.ylim(-1,1)
plt.axis('equal')
plt.axis('tight')
plt.show()
```

# 6. Labels and Legends

```python
x = np.linspace(0,10)
f = (np.sqrt(x))
g = (x**1/3)
plt.plot(x,f, label = 'f(x) = sqrt(x)')
plt.plot(x,g, label = 'g(x) = x**1/3')
plt.title('Graph of F(x) and G(x)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

# 7. Basic Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.plot(x,y, 'o')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
plt.show()
```

8. Customizing Scatter Plots

```python
x = np.linspace (0,4*np.pi)
y = (x*np.sin(x))
plt.plot(x,y, marker = 'o',
         color = 'Green',
         markerfacecolor = 'pink',
         markersize = 8,
         markeredgewidth = .5,
         linewidth = 4)
plt.show()
```

9. Combined Plot

```python
x = np.arange(-2*np.pi, 2*np.pi, np.pi/4)
y = (x*np.cos(x))
plt.plot(x,y, '-p', color = 'black',
         markerfacecolor = 'red')
plt.show()
```

# 10. Custom Plot with Alternarive Styles

I decided to graph f(x) anf h(x) since they are inverse functions and have a relationship. I decided to use five thirthy eight as an alternative style because I have never used it before and it was interesting to see how a slight difference in the background color can make certain colors pop.

```python
plt.style.use('fivethirtyeight') # changing to alternative style

x = np.linspace(-np.pi, np.pi)
f = np.tan(x)
h = (np.arctan(x))
plt.plot(x,f, '--',color = 'blue', label = 'f(x)')
plt.plot(x,h, '-',color = 'green', label = 'h(x)')
plt.legend()
plt.show()
```

# 11. Style Comparison

```python
plt.style.use('default')
x = np.random.rand(50)
y = np.random.rand(50)
plt.plot(x,y, 'o')
plt.show()
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
x = np.random.rand(50)
y = np.random.rand(50)
plt.plot(x,y, 'o')
plt.show()
```

```python
plt.style.use('fivethirtyeight')
x = np.random.rand(50)
y = np.random.rand(50)
plt.plot(x,y, 'o')
plt.show()
```

# Comparison

I believe for this graph the seaborn-whitegrid style woule be most appropriate. The default style doesn't work because it does not have grid lines which makes the graph look incomplete. Grid lines are essential when it comes to scatterplots so the viewer can tell the values of the points. Though the five-thirty-eight style and seaborn-whitegrid styles are similar, I believe the seaborn gives the graph a cleaner look and makes the points catch the eye. Meanwhile the five-thirty-eight style's background can be distracting.
