# Crime Data Visualization (2022 - 2023)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Dataset

```python
Offense = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder','Rape (revised definition)', 'Robbery']
Offenses_per_2023 = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]
Change_in_offenses_per = [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]
```

# Task 1: Basic Bar Chart for Crimes Rates in 2023

```python
Offense = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape (revised definition)', 'Robbery']
Offenses_per_2023 = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]

plt.figure(figsize=(14,8))
colors = ['#E9BBEB', '#E563EA', '#D633DC', '#B613BC', '#99379D', '#652268', '#4C0F4F']

plt.bar(Offense, Offenses_per_2023, color=colors)
plt.title('Crime Rate in 2023')
plt.xlabel('Offenses')
plt.ylabel('Offeses per 100,000 (2023)')
plt.grid(axis='y', color='gray')
plt.show()
```

# Task 2: Change in Crime Rates (2022 vs. 2023) with Unique Colors

```python
Offense = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder','Rape (revised definition)', 'Robbery']
Change_in_offenses_per = [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]

colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'pink']

plt.barh(Offense, Change_in_offenses_per, color=colors)
plt.title('Change in Crime Rates (2022 vs. 2023)')
plt.xlabel('Offenses')
plt.ylabel('Change in Offenses per 100,000 (2022-2023)')
plt.grid(axis='x', color='gray')
plt.show()
```

# Additional Instrucstions

1. Why does this bar graph make more sense than Task 1's bar graph?

Absolute numbers give less insight than percent changes. Task 1s graph only shows the crime rate for 2023. Using this graph, we dont get much insight other than which crime happens more often. However, in task 2 using  the percent change we can get more information. For example, percent changes can help with learning what caused the change between the years and what can be done in the future using that information.

# Optional Task: Combined Bar Chart for Crime Rates in 2022 and 2023

```python
Offense = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder','Rape (revised definition)', 'Robbery']
Offenses_per_2022 = [267.4, 258.8, 1352.1, 306.7, 17.7, 47.8, 67.3]
Offenses_per_2023 = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]
# Offenses_per_2022 = Offenses_per_2022 - Change_in_offenses_per

x_indexes = np.arange(len(Offense))

plt.bar(x_indexes-0.12, Offenses_per_2022,  width=0.2, color = '#A89AF8', label = '2022')
plt.bar(x_indexes+0.12, Offenses_per_2023, width = 0.2, color = '#E69AF8', label = '2023')

plt.xticks(x_indexes, Offense, rotation=45)
plt.title('Crime Rates in 2022 and 2023')
plt.xlabel('Offense')
plt.ylabel('Offense per 100,000 people')
plt.legend()
plt.show()
```
