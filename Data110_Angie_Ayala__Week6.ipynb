2. The dataset contains missing values. handle them by removing rows with NaN Values

```python
penguins=df.dropna()
penguins.info()
```

## Part 2: Data Visualization

Task 1: Boxplot - Analyzing Body Mass by Species Create a boxplot to compare the distribution of penguin body mass across different species:

```python
plt.style.use('fivethirtyeight')

sns.boxplot(x='species', y='body_mass_g', data=penguins, color=('#F89AF1'))
plt.title('Species Vs Body Mass')
plt.xlabel('Species')
plt.ylabel('Body Mass')
plt.show()
```

Interpretation: Based on the boxplot, answer the following:

2. Which species has the widest range in body mass?
3. Are there any outliers? If so, which species have them?

**Response:**
1. Gentoo has the largest median in body mass.
2. Gentoo has the widest range in body mass.
3. Yes, the Chinstrp penguins have both upper and lower outliers.

Task 2: Violin Plot - Flipper Length Distribution by Species Create a violin plot to analyze the distribution of flipper length across species:

```python
sns.violinplot(x='species', y='flipper_length_mm', data=penguins, inner='quartile', color=('#9ACEF8'), linecolor='white')
plt.title('Species Vs Flipper Length Violin Plot')
plt.xlabel('Species')
plt.ylabel('Flipper Length')
plt.show()
```

Interpretation: Answer the following:

1. Which species has the most symmetrical distribution?
2. Which species has the widest spread of flipper lengths?
3. How do the quartiles (white lines in the plot) compare for each species?

**Response:**

1. The Adelie penguin has the most symmetrical distribution.
2. The Adelie penguin has the widest spread of flipper lengths.
3. Adelie's first quartile is under Chinstrap's first quartile but Adelie's median is around where Chinstrap's first quartile is. Chinstraps quartiles lines are all under Gentoo's first quartiles.

Task 3: Ridgeline Plot - Distribution of Bill Length by Island Use a ridgeline plot to visualize the distribution of bill length across islands:

```python
data=penguins[['island', 'bill_length_mm']]

fig, axes = joyplot(penguins,
                    by='island',
                    column='bill_length_mm',
                    xlabelsize=10,
                    ylabelsize=10,
                    colormap=plt.cm.viridis)
plt.title('Island Vs Bill Lanfth Ridgeline Plot')
plt.xlabel('Bill Length', fontsize=11)
plt.show()
```

Interpretation: Answer the following:

1. Which island has the smallest range of bill lengths?
2. Which island has the most uniform distribution?
3. What differences do you notice between the islands in terms of bill length?

**Response:**
