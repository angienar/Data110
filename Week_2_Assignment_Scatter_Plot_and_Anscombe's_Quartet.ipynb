# Advanced Scatter Plot Customization

```python
plt.figure(figsize=(16,10))

plt.scatter(df['Rank'],df['HappinessScore'],
            color='blue',
            marker='o',
            edgecolors='black',
            alpha=0.75,
            s=50)

plt.xlabel('Rank', fontsize=14, fontweight='bold')
plt.ylabel('Happiness Score', fontsize=14, fontweight='bold')

plt.title('Rank Vs Happiness Score', fontsize=18, fontweight='bold')

plt.grid(True, linestyle='--', alpha=0.5)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.gca().invert_xaxis()

plt.show()
```

# Anscombe's Quartet Dataset

```python
x = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
y1 = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
y2 = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
y3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
x4 = [8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8]
y4 = [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]
```

```python
plt.scatter(x, y1)
```

```python
m, b = np.polyfit(x, y1, 1)

regression_line = m * np.array(x) + b

plt.scatter(x, y1)
plt.plot(x, regression_line,color='blue')
plt.xlabel('x')
plt.ylabel('y1')
```
